# **Melt and Merge Function of Python**

```python
#Years present in the data are present in row wise format we want it in column format
years = [str(i) for i in range(1960,2017)]
print(years)
```

```python
# Melt function on population data
df_population = pd.melt(population,
        id_vars='Country Name',
        value_vars=years,
        var_name='Year',
        value_name='Population')
df_population
```

```python
df_population.isna().sum()
```

**Merging the Country and Population data**

```python
country_and_population = pd.merge(country,df_population,how='left',on='Country Name')
```

```python
country_and_population
```

```python
# Melt function on fertility data
df_fertility = pd.melt(fertility,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='Fertility Rate')
df_fertility
```

```python
df_fertility.isna().sum()
```

```python
# Melt function on life_expectancy data
df_life_expectancy = pd.melt(life_expectancy,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='Life Expectancy')
df_life_expectancy
```

```python
df_life_expectancy.isna().sum()
```

**Merging Fertility and Life expectancy data**

```python
fertility_and_life_expectancy=pd.merge(df_fertility, df_life_expectancy, how='left', on=['Country Code', 'Year'])
fertility_and_life_expectancy
```

**Merging all four dataset i.e country, population, fertility, life_expectancy**

```python
# Merge country_and_population and fertility_and_life_expectancy
df2 = pd.merge(country_and_population, fertility_and_life_expectancy, how='left', on=['Country Code', 'Year'])
df2 = df2[['Country Name', 'Country Code', 'Region', 'Year', 'Population', 'Fertility Rate', 'Life Expectancy']]
df2
```

```python
df2.dtypes
```

```python
df2.isna().sum()
```

```python
df2.dropna(axis=0, inplace=True)
```

```python
df2.describe()
```

# **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import interact
```

**1.Creating line chart to show population trends over time w.r.to Country**

```python
countries = df2['Country Name'].unique()

# Create an empty figure
fig = go.Figure()

# Add traces for each country
for country in countries:
    country_data = df2[df2['Country Name'] == country]
    fig.add_trace(go.Scatter(x=country_data['Year'], y=country_data['Population'], mode='lines+markers', name=country))

fig.update_layout(
    title='Population Trends Over Time',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Population'),
    showlegend=True,
    updatemenus=[
        {
            'buttons': [
                {
                    'method': 'update',
                    'label': country,
                    'args': [{'y': [df2[df2['Country Name'] == country]['Population']], 'name': country}]
                } for country in countries
            ],
            'direction': 'down',
            'showactive': True,
            'x': 0.5,
            'xanchor': 'center',
            'y': 1.15,
            'yanchor': 'top'
        }
    ]
)

# Show the plot
fig.show()
```

**Created filter on  Country to show population trends over time w.r.to Country**

```python
def plot_population(country):
    country_data = df2[df2['Country Name'] == country]
    plt.figure(figsize=(22, 6))  # Adjust figure size to accommodate the longer x-axis labels
    plt.plot(country_data['Year'], country_data['Population'],color='magenta', marker='o', linestyle='-', label=country)
    plt.title('Population Change Over Time')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.legend(loc='upper left')  # Change the legend position if needed
    plt.xticks(rotation=50)  # Rotate x-axis labels for better readability
    plt.grid(True)
    plt.show()

# Get unique countries
countries = df2['Country Name'].unique()

# Create interactive plot using ipywidgets
interact(plot_population, country=countries);
```

**Created filter on Country to show Life Expectancy trends over time w.r.to Country**

```python
def life_expectancy_rate(country):
    country_data = df2[df2['Country Name'] == country]
    plt.figure(figsize=(22, 6))
    plt.plot(country_data['Year'], country_data['Life Expectancy'],color='green', marker='o', linestyle='-', label=country)
    plt.title('Life Expectancy Rate Change Over Time')
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy Rate')
    plt.legend(loc='upper left')  # Change the legend position if needed
    plt.xticks(rotation=50)  # Rotate x-axis labels for better readability
    plt.grid(True)
    plt.show()

# Get unique countries
countries = df2['Country Name'].unique()

# Create interactive plot using ipywidgets
interact(life_expectancy_rate, country=countries);
```

**Created filter on Country to show Fertility Rate trends over time w.r.to Country**

```python
def fertility_rate(country):
    country_data = df2[df2['Country Name'] == country]
    plt.figure(figsize=(22, 6))
    plt.plot(country_data['Year'], country_data['Fertility Rate'],color='red', marker='o', linestyle='-', label=country)
    plt.title('Fertility Rate Rate Change Over Time')
    plt.xlabel('Year')
    plt.ylabel('Fertility Rate')
    plt.legend(loc='upper left')  # Change the legend position if needed
    plt.xticks(rotation=50)  # Rotate x-axis labels for better readability
    plt.grid(True)
    plt.show()

# Get unique countries
countries = df2['Country Name'].unique()

# Create interactive plot using ipywidgets
interact(fertility_rate, country=countries);
```

```python
region_population = df2.groupby('Region')['Population'].sum().reset_index()

# Define colors for each region
colors = ["orange", "red", "green", "blue", "purple",'magenta','cyan']  # Example colors, you can define your own

# Create a Plotly bar plot with customized colors
fig = go.Figure(go.Bar(
    x=region_population['Region'],
    y=region_population['Population'],
    marker_color=colors
))

# Update layout
fig.update_layout(
    title='Population by Region',
    xaxis=dict(title='Region'),
    yaxis=dict(title='Population'),
    showlegend=False
)

# Show the plot
fig.show()
```
